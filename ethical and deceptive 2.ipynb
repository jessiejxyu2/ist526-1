```python
filtered_data = df[df['Variable'] == 'COVID-19 deaths (number)']
filtered_data
```

```python
# Filter the DataFrame for the desired criteria (replace 'COUNTRY', 'GENDER', 'YEAR', and 'Value' as needed)
filtered_data = df[(df['COUNTRY'] == 'USA') & (df['VARIABLE'] == 'COVIDNB')]

# Create a continuous index combining year and week
filtered_data['YearWeek'] = (filtered_data['YEAR'] - filtered_data['YEAR'].min()) * 52 + filtered_data['WEEK']

# Group the data by 'YearWeek' and calculate the sum of the 'Value' column
COVID19_deaths = filtered_data.groupby('YearWeek')['Value'].sum()

# Create the line chart
fig, ax = plt.subplots()
COVID19_deaths.plot(kind='bar', ax=ax)
ax.set_ylabel('number of COVID19 deaths by week')

# Set x-axis labels with 6-month intervals and their corresponding years
num_weeks = len(COVID19_deaths)
interval = 26  # 6-month interval in weeks
xticks = np.arange(0, num_weeks, interval)
xticklabels = [(filtered_data['YEAR'].min() + int(week / 52)) for week in xticks]
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels, rotation=45)
plt.show()
```

```python
# Filter data for the years 2020-2021
filtered_data = filtered_data[(filtered_data['YEAR'] >= 2020) & (filtered_data['YEAR'] <= 2021)]

# Create a continuous index combining year and week
filtered_data['YearWeek'] = (filtered_data['YEAR'] - filtered_data['YEAR'].min()) * 52 + filtered_data['WEEK']

# Group the data by 'YearWeek' and calculate the sum of the 'Value' column
COVID19_deaths = filtered_data.groupby('YearWeek')['Value'].mean()

# Create the bar chart
fig, ax = plt.subplots()
COVID19_deaths.plot(kind='bar', ax=ax)
ax.set_ylabel('number of COVID 19 deaths by week')
ax.set_xlabel('  ')
# Set x-axis labels with 3-month intervals and their corresponding years
num_weeks = len(COVID19_deaths)
interval = 13  # 3-month interval in weeks
xticks = np.arange(0, num_weeks, interval)
xticklabels = ['2020 Jan.', '2020 Mar.', '2020 Jun.', '2020 Oct.', '2021 Jan.', '2021 Mar.', '2021 Jun.', '2021 Oct.']
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels, rotation=45)

plt.show()
```

```python
# Filter data for the year 2020
filtered_data = filtered_data[(filtered_data['YEAR'] == 2020)]

# Group the data by 'YearWeek' and calculate the sum of the 'Value' column
COVID19_deaths = filtered_data.groupby('YearWeek')['Value'].mean()

# Create the bar chart
fig, ax = plt.subplots(figsize=(15, 6))  # Adjust the figure size here
COVID19_deaths.plot(kind='bar', ax=ax)
ax.set_ylabel('number of COVID 19 deaths by week')
ax.set_xlabel('week in year 2020')
# Rotate the x-axis labels
xticks = ax.get_xticks()
xticklabels = ax.get_xticklabels()
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels, rotation=0)
ax.set_title("USA Covid 19 deaths by week in 2020")
plt.show()
```
